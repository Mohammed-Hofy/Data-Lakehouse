# Read data from CSV files and write them as Bronze tables using overwrite mode

```python
df = spark.read.csv('/Volumes/workspace/bronze/source_systems/source_crm/cust_info.csv',
                    header=True ,inferSchema=True)

df.write.mode('overwrite').saveAsTable('bronze.crm_cust_info')
```
