# Gold Layer – Sales Fact

This notebook builds the sales fact table by combining sales transactions with customer and product dimensions to support business reporting and analytics.

## Build Sales Fact Table

Combine sales transactions with customer and product dimension keys to create the sales fact table.

```python
df =  spark.sql( """
SELECT 
    s.order_number, 
    p.product_key,  
    c.customer_key,  
    s.order_date,
    s.ship_date, 
    s.due_date, 
    s.sales_amount, 
    s.quantity, 
    s.price
FROM workspace.silver.crm_sales s
LEFT JOIN workspace.gold.dim_customers c
ON s.customer_id = c.customer_id
LEFT JOIN workspace.gold.dim_products p
ON s.product_number = p.category_id
"""
)
```

## Preview Fact Data

Review the sales fact table before loading it into the Gold layer.

```python
df.limit(10).display()
```

## Load to Gold Layer

Write the sales fact table to the Gold layer.

```python
df.write.mode('overwrite').saveAsTable('workspace.gold.fact_sales')
```

## Verify Data Load

Validate that the sales fact table has been successfully loaded into the Gold layer.

```python
%sql
SELECT * 
FROM workspace.gold.fact_sales LIMIT 10
```
