# Silver Layer – Product Category Information

This notebook prepares product category data for the Silver layer by applying column standardization and business-friendly naming conventions.

```python
from pyspark.sql.functions import *
df = spark.read.table('workspace.bronze.erp_px_cat_g1v2')
```

## Rename Columns

Rename columns to follow the Silver layer naming convention.

```python
RENAME_MAP = {
    "id": "category_id",
    "cat": "category",
    "subcat": "subcategory",
    "maintenance": "maintenance_flag"
}
for old_name, new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```

## Preview Transformed Data

Review the dataset before loading it into the Silver layer.

```python
df.limit(10).display()
```

## Load to Silver Layer

Write the product category data to the Silver layer.

```python
df.write.mode('overwrite').saveAsTable('workspace.silver.erp_product_category')
```

## Verify Data Load

Validate that the product category data has been successfully loaded into the Silver layer.

```python
%sql
SELECT *
FROM workspace.silver.erp_product_category LIMIT 10
```
